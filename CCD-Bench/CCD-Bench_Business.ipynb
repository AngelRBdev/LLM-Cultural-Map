```python
corporateEntities = ['business', 'company', 'organization', 'workplace', 'workspace']
hierarchy = ['employer', 'employee', 'boss', 'leader', 'manager', 'colleague', 'coworker', 'co-worker']
career = ['career', 'job', 'promotion', 'appraisal', 'professional']
operations = ['team', 'project', 'task', 'overtime']

workplace_terms = corporateEntities + hierarchy + career + operations

mcq_business = pd.concat([
    mcq[
        mcq['Question'].str.contains('|'.join(workplace_terms), case=False, na=False) +
        mcq['Domain'].str.contains('Business', case=False, na=False)
    ]
])

mcq_business.count()
```

```python
mcq_business[mcq_business.duplicated()].count()
```

```python
to_drop = culturalDilemmas.columns[culturalDilemmas.columns.str.contains('rationale')].tolist() + ["Question", "Domain"]

to_drop
```

```python
mcq_business = pd.concat([mcq_business, culturalDilemmas.drop(to_drop, axis=1)],
                         axis=1,
                         join='inner')

mcq_business.info()
```

## The Remaining 130+ UN-recognized Countries

The world has 195 UN-recognized countries, so for the rest of the remaining societies we mapped according to the following rationale:

1. **Anglo:** Extended to include English-speaking Caribbean nations and former British colonies in the Americas that share institutional and linguistic heritage with the UK and US.

2. **Confucian Asia:** Extended to East Asian nations heavily influenced by Confucian philosophies, primarily Vietnam and North Korea.

3. **Eastern Europe:** Extended to the remaining Slavic, Balkan, Baltic, and Caucasus nations, as well as Central Asian former Soviet republics (which share institutional histories similar to Kazakhstan and Georgia, both originally in this cluster).

4. **Germanic Europe:** Extended to Luxembourg and Liechtenstein, which share high uncertainty avoidance, performance orientation, and Germanic linguistic roots.

5. **Latin America:** Extended to the remaining Spanish, Portuguese, and French-speaking nations in the Americas and the Caribbean (e.g., Chile, Peru, Cuba).

6. **Latin Europe:** Extended to the remaining Romance-language speaking European micro-states (e.g., Andorra, Monaco) and Malta. While Belgium is linguistically split, its institutional structure often aligns it with Latin Europe in extended cultural studies.

7. **Middle East (MENA):** Extended to the remaining Arab nations in the Levant, Gulf, and North Africa (e.g., Saudi Arabia, UAE, Algeria, Syria) sharing strong in-group collectivism and Islamic institutional influence.

8. **Nordic Europe:** Extended to Norway and Iceland, which share the exceptionally high gender egalitarianism and institutional collectivism of Sweden, Denmark, and Finland.

9. **Southern Asia:** Extended to the remaining South and Southeast Asian countries (e.g., Pakistan, Bangladesh, Cambodia). Pacific Island nations (Oceania) are also placed here due to their Austronesian ethnolinguistic roots and similar tribal/in-group collectivism profiles, though they remain an under-researched geographic block.

10. **Sub-Saharan Africa:** Extended to the remaining African nations south of the Sahara, which share high humane orientation and extended community welfare structures.
