```python
import pandas as pd
mcq = pd.read_json('CDEval.json')
culturalDilemmas = pd.read_json('CCD-Bench.json')
globe = pd.read_json('GLOBE.json')
```

```python
corporateEntities = ['business', 'company', 'organization', 'workplace', 'workspace']
hierarchy = ['employer', 'employee', 'boss', 'leader', 'manager', 'colleague', 'coworker', 'co-worker']
career = ['career', 'job', 'promotion', 'appraisal', 'professional']
operations = ['team', 'project', 'task', 'overtime']

workplace_terms = corporateEntities + hierarchy + career + operations

mcq_business = pd.concat([
    mcq[
        mcq['Question'].str.contains('|'.join(workplace_terms), case=False, na=False) +
        mcq['Domain'].str.contains('Business', case=False, na=False)
    ]
])

mcq_business.count()
```

```python
mcq_business[mcq_business.duplicated()].count()
```

```python
to_drop = culturalDilemmas.columns[culturalDilemmas.columns.str.contains('rationale')].tolist() + ["Question", "Domain"]

to_drop
```

```python
mcq_business = pd.concat([mcq_business, culturalDilemmas.drop(to_drop, axis=1)],
                         axis=1,
                         join='inner')

mcq_business.info()
```
